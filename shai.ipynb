```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv("Salaries.csv")
df.head()
```

```python
df.columns
```

```python
#1- Identify the number of rows and columns in the dataset
num_rows, num_columns = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
```

```python
# Determine the data types of each column
df.dtypes
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

```python
#2-Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]  
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()

salary_range = max_salary - min_salary

std_dev_salary = df['TotalPay'].std()

print(f"Mean Salary: {mean_salary:.2f}")
print(f"Median Salary: {median_salary:.2f}")
print(f"Mode Salary: {mode_salary:.2f}")
print(f"Minimum Salary: {min_salary:.2f}")
print(f"Maximum Salary: {max_salary:.2f}")
print(f"Salary Range: {salary_range:.2f}")
print(f"Standard Deviation of Salary: {std_dev_salary:.2f}")
```

```python
df.sample(10)
```

```python
#3-Data Cleaning: Handle missing data by suitable method with explain why you use it.
df = df.drop(['Status', 'Notes','Benefits'], axis=1)
df.sample(10)
#here i delete three columns, status and notes because all of it have a non value, and benefits because half of it has a non value
```

```python
# Remove rows with any null values
df_cleaned = df.dropna()

df_cleaned.sample(10)
```

```python
# Drop rows with 0 values in any column
df_cleaned = df[(df != 0).all(axis=1)]
df_cleaned.sample(10)
```

```python
df_cleaned.shape
```

```python
#4-Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.


import matplotlib.pyplot as plt

# Set the style for plots (optional)
plt.style.use('seaborn')

# Create a histogram for the distribution of salaries
plt.figure(figsize=(10, 6))
plt.hist(df_cleaned['TotalPay'], bins=30, edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart representing the proportion of employees in different departments
department_counts = df_cleaned['JobTitle'].value_counts().head(10)  # Adjust the number of departments to display
labels = department_counts.index
sizes = department_counts.values

plt.figure(figsize=(10, 10))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
# Assuming 'JobTitle' column is used for the grouped analysis
top_job_titles = df_cleaned['JobTitle'].value_counts().head(20).index

# Group the data by 'JobTitle' and calculate average salaries for each group
grouped_data = df_cleaned[df_cleaned['JobTitle'].isin(top_job_titles)].groupby('JobTitle')['TotalPay'].agg(['count', 'mean', 'median', 'min', 'max'])


grouped_data
```

```python
# Compare average salaries across different job titles (Bar chart)
plt.figure(figsize=(12, 6))
grouped_data.sort_values('mean', ascending=False)['mean'].plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Salaries Across Different Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.show()
```

```python
import seaborn as sns

# Choose another numerical column for correlation analysis 
numerical_column = 'BasePay'

# Check the correlation between 'TotalPay' and the selected numerical column
correlation = df_cleaned['TotalPay'].corr(df_cleaned[numerical_column])

# Visualize the relationship with a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=numerical_column, y='TotalPay', data=df_cleaned, color='skyblue')
plt.title(f'Scatter Plot: TotalPay vs {numerical_column} (Correlation: {correlation:.2f})')
plt.xlabel(numerical_column)
plt.ylabel('TotalPay')
plt.show()
```
